```python
import pandas as pd
import wget
from zipfile import ZipFile
import plotly.graph_objects as go

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

#### Acessando banco de dados e criando arquivos históricos

```python
url = 'http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/DFP/DADOS/'
```

```python
# Criando uma lista com o nome de todos os arquivos
arquivos_zip = []

for ano in range(2010,2023):
    arquivos_zip.append(f'dfp_cia_aberta_{ano}.zip')
```

```python
arquivos_zip
```

```python
#Fazendo dowload de todos os arquivos

for arq in arquivos_zip:
    wget.download(url+arq)
```

```python
# extraindo os arquivos zip

for arq in arquivos_zip:
    ZipFile(arq, 'r').extractall('CVM_DFP')
```

```python
import yfinance as yf
```

```python
itub4 = yf.download("ITUB4.SA", start='2011-01-01')
```

```python
itub4.head()
```

```python
indicadores = itub4.join(empresa['VL_CONTA'], how='outer')
```

```python
indicadores.rename({'VL_CONTA':'LPA'}, axis= 1, inplace=True)
```

```python
indicadores.head()
```

```python
indicadores.fillna(method='ffill', inplace=True)
```

```python
indicadores.dropna(inplace=True)
```

```python
indicadores.head()
```

```python
indicadores['PL'] = indicadores['Close'] / indicadores['LPA']
indicadores['PL Adj'] = indicadores['Adj Close'] / indicadores['LPA']
```

```python
indicadores.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=indicadores.index, y=indicadores['PL'], name='PL'))
fig.add_trace(go.Scatter(x=indicadores.index, y=indicadores['PL Adj'], name='PL Adj'))
fig.add_trace(go.Scatter(x=indicadores.index, y=indicadores['Close'], name='ITUB4'))
fig.add_trace(go.Scatter(x=indicadores.index, y=indicadores['PL'], name='ITUB4 Adj'))
```
